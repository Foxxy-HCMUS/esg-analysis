```python
new_df = df.copy()
# Extract the year values from the column headers
year_cols = [col for col in new_df.columns if col.isnumeric()]

# Define the columns to keep as identifiers
id_vars = ['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code']

# Pivot the value columns into a single column
new_df = new_df.melt(id_vars=id_vars, value_vars=year_cols, var_name='Year', value_name='Value')

# Display the pivoted data
display(new_df)
```

## Lọc dữ liệu region và 1 số aggregation ra 1 file khác

```python
countries_df = pd.read_csv('..\\data\\raw\\ESGCountry.csv')
```

- Các nước không có đơn vị tiền tệ:

```python
countries_df['Special Notes'].str.contains('aggregate', na=False).sum()
countries_df['Currency Unit'].isna().sum()
```

- Các nước không có đơn vị tiền tệ là các group aggregation trong dataset, ta sẽ lọc các giá trị này ra 1 dataset khác

```python
countries_code_sr = countries_df[countries_df['Currency Unit'].notna()]['Country Code']
countries_code_sr
```

```python
agg_df = new_df[~new_df['Country Code'].isin(countries_code_sr)]
agg_df = agg_df[agg_df['Value'].notnull()]

grouped = agg_df.groupby(['Country Name', 'Indicator Name']).agg({'Value': 'count'})
grouped = grouped[grouped['Value'] > 0]
grouped = grouped.reset_index()
agg_df = pd.merge(agg_df, grouped[['Country Name', 'Indicator Name']], on=['Country Name', 'Indicator Name'])

agg_df
```

```python
series_df = pd.read_csv('..\\data\\raw\\ESGSeries.csv')
field_df = series_df[['Series Code', 'Topic']]
agg_df = pd.merge(agg_df, field_df, left_on='Indicator Code', right_on='Series Code', how='left')
agg_df.drop(['Series Code', 'Indicator Code'], axis=1, inplace=True)
agg_df.to_csv('..\\data\\processed\\ESGData_agg.csv')
```

## Chọn ra các dữ liệu thuộc Social
